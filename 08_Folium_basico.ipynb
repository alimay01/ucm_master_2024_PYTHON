```python
m = folium.Map(location = coordenadas_ucm_docu,
    zoom_start=14, 
    tiles='Stamen Terrain' 
) 
tooltip = 'UCM docu'
folium.Marker(coordenadas_ucm_docu, 
              popup = 'UCM', 
              tooltip = tooltip).add_to(m) 
```

```python
m
```

```python
m = folium.Map(location = coordenadas_ucm_docu,
    zoom_start=14 
) 
tooltip = 'UCM docu'
folium.Marker(coordenadas_ucm_docu, 
              popup = 'UCM', 
              tooltip = tooltip).add_to(m) 
```
